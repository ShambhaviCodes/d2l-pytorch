# Image Augmentation

Image augmentation technology expands the scale of training datasets by making a series of random changes to the training images to produce similar, but different, training examples. Another way to explain image augmentation is that randomly changing training examples can reduce a model's dependence on certain properties, thereby improving its capability for generalization. For example, we can crop the images in different ways, so that the objects of interest appear in different positions, reducing the model's dependence on the position where objects appear. We can also adjust the brightness, color, and other factors to reduce model's sensitivity to color. It can be said that image augmentation technology contributed greatly to the success of AlexNet. In this section, we will discuss this technology, which is widely used in computer vision.

Let's start with importing necessary libraries.

```
import PIL
from PIL import Image
import numpy as np
import torch
import torchvision
```

We will apply the following transforms to an image using 'torchvision.transforms' :


*   Resize the image to (224, 224)
*   Hue and Saturation to (0.05)
*   Horizontal Flip
*   Rotation





```
transforms = torchvision.transforms.Compose([
    torchvision.transforms.Resize((224,224)),
    torchvision.transforms.ColorJitter(hue=.05, saturation=.05),
    torchvision.transforms.RandomHorizontalFlip(),
    torchvision.transforms.RandomRotation(20, resample=PIL.Image.BILINEAR)
])
```

```
img = Image.open("/content/drive/My Drive/Dataset/7.jpg")
```

```
transforms(img)
```

We can see the augmentation in the image. These transforms are very basic, one may utilise libraries such as [imgaug](https://github.com/aleju/imgaug) for more interesting transformations.

## Summary 
Image Augmentation is the process of generating new images for training our deep learning model. It helps in expanding the dataset and helps in a more robust training to perform more accurately.
